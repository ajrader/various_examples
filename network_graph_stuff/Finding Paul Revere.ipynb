## Calculate the adjaceny matrices for groups (Ag) and people (Ap)

```python
A =revere[['StAndrewsLodge','LoyalNine','NorthCaucus','LongRoomClub','TeaParty','BostonCommittee','LondonEnemies']]
```

```python
Ag = dot((A.values.T),(A.values))
```

```python
shape(Ag)
```

```python
Ap = dot((A.values),(A.values.T))
shape(Ap)
```

## Push these into 2 Graphs using Networkx module

```python
print Ag
groupNames
```

```python
#remove the diagonals to eliminate self-loops & create a version that is unweighted
Agg = Ag - diag(Ag)*eye(len(Ag))
print Agg
Ugg = (Agg > 0).astype(int)
```

```python
group = nx.Graph()
group.add_nodes_from(groupNames)
group.nodes()
```

```python
for i in xrange(0,len(group)):
    for j in xrange(i+1,len(group)):
        wtval = Ag[i,j]
        group.add_edge(groupNames[i],groupNames[j],weight=wtval)
```

```python
#show the adjacency matrix
group.adj
```

## Redraw with node size proportional to # members and edge width to #shared members

```python
#push weights into a list
edgewidth=[]
for (u,v) in group.edges():
    #print group.get_edge_data(u,v)['weight']
    wt = group.get_edge_data(u,v)['weight']
    edgewidth.append( wt )

print edgewidth
#push membership size into an array
nodesize = []
for i in xrange(0,len(group)):
    nsize= Ag[i,i]
    nodesize.append(nsize*10+300)

print nodesize
```

```python
# function to write a gexf file from adjacency matrix
def adj2gexf_fcn(App,labels,fname,author,descript,directed,thickflag):
    import datetime as dt
    #import argparse.ArgurmentParser(description='Convert an adjaceny matrix to gexf file')
    #parser = argparse
    today = dt.date.today()
    nodes = len(App)
    kedge = 0
    print today, nodes, fname
    etype ={0:'undirected', 1:'directed'}
    
    #f = open(fname,'w')
    headerinfo =['<?xml version="1.0" encoding="UTF-8"?>\n',
                 '<gexf xmlns="http://www.gexf.net/1.2draft"\n',               
                 '\t\t\t xmlns:xsi="http://www.w3.org/2001/XMLSchemainstance"\n',
                 '\t\t\t xsi:schemaLocation="http://www.gexf.net/1.2draft\n',
                 '\t\t\t\t\t http://www.gexf.net/1.2draft/gexf.xsd"\n',
                 '\t\tversion="1.2">\n',
                 '\t<meta lastmodifieddate="'+str(today)+'">\n',
                 '\t\t<creator>'+str(author)+'</creator>\n',
                 '\t\t<description>'+str(descript)+'</description>\n',
                 '\t</meta>\n',
                 '\t<graph defaultedgetype="'+str(etype[directed])+'">\n',
                 '\t\t<nodes>\n']
    f = open(fname,'a')
    for a in headerinfo:
        f.write(a)
    
    
    for i in xrange(0,nodes):
        stemp = '\t\t\t<node id="'+str(i)+'" label="'+str(labels[i])+'"/>\n'
        f.write(stemp)
    
    f.write('\t\t</nodes>\n')
    f.write('\t\t<edges>\n')
    
    for i in xrange(0,nodes):
        for j in xrange(i+1,nodes):
            wtval = App[i,j]
            if(wtval !=0):
                stemp ='\t\t\t<edge id ="'+str(kedge)+'" source = "'+str(i)+'" target="'+str(j)+'" weight="'+str(wtval)+'"/>\n'
                f.write(stemp)
                if thickflag != 0 :
                    thickline = '\t\t\t\t<vis:thickness value="'+str(wtval)+'"/>\n'
                    f.write(thickline)
                kedge=kedge+1
            
    f.write('\t\t</edges>\n')
    f.write('\t</graph>\n')
    f.write('</gexf>\n')    
    
    f.close()
    print " %s edges" % kedge
```
