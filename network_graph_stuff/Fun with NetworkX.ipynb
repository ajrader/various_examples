<H2> Use Dijkstra's algorithm to find shortest path in a weighted and unweighted  network:</H2>

```python
import networkx as nx
```

```python
g = nx.Graph() # create an empty graph
# add nodes a,b,c & d with edges between various components
ww = [0.1,1.5,1.0,2.2]
g.add_edge('a','b',weight =ww[0])
g.add_edge('b','c',weight =ww[1])
g.add_edge('a','c',weight =ww[2])
g.add_edge('c','d',weight =ww[3])
```

```python
# plot the graph
nx.draw(g)
```

```python
nodeXY = nx.spring_layout(g)
edgewidth=[]
for (u,v) in g.edges():
    wt = g.get_edge_data(u,v)['weight']
    print u,v,wt
    edgewidth.append(wt*4)
                
nx.draw_networkx_nodes(g,pos=nodeXY,node_color='antiquewhite')
nx.draw_networkx_labels(g,nodeXY,fontsize=14)
nx.draw_networkx_edges(g,nodeXY,node_size=0,width=edgewidth,edge_color='darkblue')
```

```python
print nx.shortest_path(g,'d','b')
print nx.shortest_path(g,'d','b',weight=True) 
```

<H3> Load the Hartford Drug Users graph information: This is a directed graph with integers as nodes.</H3>

```python
hartford = nx.read_edgelist('../../../../data/miscDataSets/hartford_drug.txt',
                            create_using=nx.DiGraph(),nodetype=int)
```

```python
N,K = hartford.order(), hartford.size()
avg_deg = float(K)/N
print "Nodes: ", N
print "Edges: ", K
print "Average degree: ", avg_deg
```

<h3> Compute in and out degree distributions of the graph and plot them (may be problematic for large graphs</h3>

```python
in_degrees = hartford.in_degree()
in_values = sorted(set(in_degrees.values()))
in_hist = [in_degrees.values().count(x) for x in in_values]
# now for out degrees:
out_degrees = hartford.out_degree()
out_values = sorted(set(out_degrees.values()))
out_hist = [out_degrees.values().count(x) for x in out_values]
```

```python
figure()
plot(in_values,in_hist,'o-',color='coral') # in-degree
plot(out_values,out_hist,'v-',color='cadetblue') # out-degree
legend(['In-degree','Out-degree'])
xlabel('Degree')
ylabel('Number of nodes')
title('Hartford drug users network')
#savefig('hartford_degree_distribution.pdf')
#plt.close()
```
