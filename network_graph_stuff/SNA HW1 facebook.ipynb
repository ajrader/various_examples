```python
#Use networkX
import networkx as nx
filename = 'fb.gml'
#filename = 'LadaFacebookAnon.gml'
```

### Read the graph using networkX

```python
#G = nx.read_gml(filename,relabel=True) # this failed 
G = nx.read_gml(filename)
```

#### Find number of nodes:

```python
num_nodes = G.number_of_nodes()
print 'number of nodes: ' + str(num_nodes)
```

```python
num_edges = G.number_of_edges()
print 'number of edges: ' +str(num_edges)
```

```python
G_connected_components = nx.connected_components(G)
num_connected_components = len(G_connected_components)
print 'number of connected components: ' + str(num_connected_components)
```

```python
max_degree_size = max(G.degree().values())
print 'maximum degree: '+str(max_degree_size)
```
