```python
import pandas.rpy.common as com
```

```python
AP = com.load_data('AirPassengers')
```

```python
AP.head()
```

```python
%matplotlib inline
import matplotlib.pylab as plt
```

```python
AP.plot()
```

```python
import pandas as pd
```

```python
pd.date_range?
```

```python
ts_start = AP.index[0]
ts_end = AP.index[-1]
ts_freq = 'M'
ts_length = len(AP.index)
my_ts = pd.date_range(str(int(ts_start)),periods=ts_length,freq=ts_freq)
my_ts
```

```python
AP.index = my_ts
AP.plot()
```

```python
AP.resample('A',how=sum).plot()
```
