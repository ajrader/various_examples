## Why do I care?

### DTW is a specific application of a dynamic programming algorithm for pairwise alignment.
* Dynamic Programming is cool $\to$ basically it means solving a complex problem by breaking it down into a set of simpler problems.
* Widely used in time series, bioinformatics (sequence alignment), speech recognition, handwriting, etc. (see [wikipedia](http://en.wikipedia.org/wiki/Dynamic_programming) for examples)
* provides a metric to find the optimal alignment between two signals. 
* time is allowed to be 'warped', with certain constraints, in order to create this best alignment.
* then these signals can be clustered together based upon their similarity (DTW distance).


<hr>
### First a little math...
* Starting with two sequences, $Q = q_1,q_2,...,q_n$ and $C = c_1,c_2,...,c_m$, alignment using DTW requires constructing the $n$ by $m$ matrix, ${\cal D}$, where the element ${\cal D}_{ij}  = d(q_i,q_j) = (q_i - c_j)^2$ is the distance between these $i^{th}$ and $j^{th}$ elements.
* The optimal path is then the one that minimizes the "warping cost": $ DTW(Q,C) = \min(\sqrt(\sum_{k=1}^K w(i,j)_k))$ where $w(i,j)_k$ is the warping cost for the aligned pair: $(i^{th},j^{th})$.

## Goal is to identify an optimal mapping between all points in the 2 signals.
### Step 1 is to calculate how far apart these signals are.
* construct the 2D matrix of distances between all pairs in signals q and c

````python

# assign Euclidean distances:
def assign_euclidean_dist(x,y):
    # initialize
    distances = np.zeros((len(y),len(x)))
    for i in xrange(0,len(y)):
        for j in xrange(len(x)):
            distances[i,j]= (q[j]-y[i])**2
    return distances

# define a function to visualize the distance matrix
def distance_cost_plot(distances,color_map='Reds'):
    im = plt.imshow(distances, interpolation='nearest', cmap=color_map) 
    plt.gca().invert_yaxis()
    plt.xlabel("Q")
    plt.ylabel("C")
    plt.grid()
    plt.colorbar();
````

````python
my_dist = assign_euclidean_dist(q,c)
print "It looks like the diagonals (or near to it have the lowest distances)"
distance_cost_plot(my_dist)
````

## Define the accumulated cost along a pathway (this is the recursion step)

````python
def calculate_accumulated_cost(x,y,distances):
    # initialize
    accumulated_cost = np.zeros((len(y),len(x)))
    # corner value is same
    accumulated_cost[0,0]= distances[0,0]
    # distance_cost_plot(accumulated_cost)
    # assign initial row
    for i in range(1, len(x)):
        accumulated_cost[0,i] = distances[0, i] + accumulated_cost[0,i-1]   
    #distance_cost_plot(accumulated_cost)
    # assign initial column
    for i in range(1, len(y)):
        accumulated_cost[i,0] = distances[i, 0] + accumulated_cost[i-1, 0]   
    #distance_cost_plot(accumulated_cost)
    # fill out the rest of the matrix
    for i in range(1, len(y)):
        for j in range(1, len(x)):
            accumulated_cost[i, j] = min(accumulated_cost[i-1, j-1], accumulated_cost[i-1, j], accumulated_cost[i, j-1]) + distances[i, j]

    return accumulated_cost
````

## Identify the optimal (warping) path via Traceback/Backtracking.

````python
# wrap this into a function
def path_cost(x, y, accumulated_cost, distances):
    path = [[len(x)-1, len(y)-1]]
    cost = 0
    i = len(y)-1
    j = len(x)-1
    while i>0 and j>0:
        if i==0:
            j = j - 1
        elif j==0:
            i = i - 1
        else:
            if accumulated_cost[i-1, j] == min(accumulated_cost[i-1, j-1], accumulated_cost[i-1, j], accumulated_cost[i, j-1]):
                i = i - 1
            elif accumulated_cost[i, j-1] == min(accumulated_cost[i-1, j-1], accumulated_cost[i-1, j], accumulated_cost[i, j-1]):
                j = j-1
            else:
                i = i - 1
                j= j- 1
        path.append([j, i])
    path.append([0,0])
    for [y, x] in path:
        cost = cost +distances[x, y]
    return path, cost    
````

# The Myths
## Myth 1.
### The ability of DTW to handle sequences of different lengths is a great advantage, and therefore the simple lower bound that requires different-length sequences to be reinterpolated to equal length is of limited utility.
## Myth 2.
### Constraining the warping paths is a necessary evil that we inherited from the speech processing community to make DTW tractable, and that we should find ways to speed up DTW with no (or larger) constraints.
## Myth 3. 
### There is a need (and room) for improvements in the speed of DTW for data mining applications.

### Maximum Warping Window that is Allowed
* Basically to compute the full distance matrix between two signals (A and B here) one needs $O(nm)$ calculations where $n$ and $m$ are the lengths of A and B.
* This isn't necessarily an issue if you are just doing the comparison between 2 signals, but what if you have p signals?
    - ${p \choose 2} = \frac{p!}{2!(p-2)!}$
    - so p=1000 means 499,500 * n * m calculations are required.
* Restricting the amount of warping allowed (Sakoe-Chiba Band or Itakura Parallelogram) reduces the number of elements that have to be calculated (and stored).

>"If we are willing to force the sequences to be of the same length, and to constrain the warping, then we have a simple solution for speeding up similarity search under DTW."

### Section 3 addresses the first myth: do we need to reinterpolate different length sequences or not?
* Figure 4 shows how indistinguishable the accuracies are for 4 different datasets using different-length sequences and reinterpolated sequences. 
* interesting comments:
    > Note that the experiments do strongly suggest that changing the amount of warping allowed does affect the accuracy.
    > 
    > we were unable to create an artificial problem where reinterpolating made a significant difference in accuracy. 

### Section 4 addresses the 2nd myth: tight constraints in warping window is a necessary evil that "throws out the baby with the bathwater".
* Figure 5 shows that "wider warping constraints do not always improve the accuracy, as commonly believed".
* "... there is no evidence to support the idea that we need to support wider constraints" 
* typically the 10% warping window constraint used by default in speech processing applications is too large for most real world data mining approaches.

### Section 5 deals with the 3rd myth: DTW needs further speedups.
* "... for problems involving a few thousand sequences or more, each with a few hundred data points, the _significant CPU cost of DTW_ is simply a non-issue."
* "it is almost certainly pointless to attempt to speed up the CPU time for DTW by producing tighter lower bounds." 
    - only marginal improvements have been observed, negating the cost of implementation.

### Using DP or DTW for sequence alignment, is the connection to last week's NLP conversation.
* bioinformatics sequence alignment

##Aside on (pair-wise) sequence alignment 
* (Needleman_Wunsch approach [1970]) 
    - Great for optimal, i.e. global alignment of two sequences
    - detailed description can be found [here](http://en.wikipedia.org/wiki/Needleman%E2%80%93Wunsch_algorithm)

### Steps
1. Consider two sequences: $A =$ ```GCATGCU``` and $B=$ ```GATTACA```
2. Arrange these sequences (vectors) along the axes of a 2D matrix 
3. Score the _distance_ between each element of this matrix 
    - various scoring methods can be used, but the dynamic programming part relies upon using a recursive calculation that includes the path to get to that element ($\cal{D}_{ij}$). 
    - if you allow gaps, i.e. ```GAT-TACA```, then $g$ is the penalty for making such an insertion.
    - for the first row and column assign the values $\cal{D}_{0j} = g*j$ and $\cal{D}_{i0} = g*i$.
    - The optimal score for a given element is given by solving: ${\cal D}_{ij} = \max({\cal D}_{i-1,j-1}+s(A_i,B_j), \cal{D}_{i,j-1}+g, \cal{D}_{i-1,j}+g)$ where $s(A_i,B_j)$ is a measure of how similar $A_i$ and $B_j$ are.
4. Traceback. Find the best alignment using the scores stored in the matrix ${\cal D}$.
    - start in the lower right corner of the matrix (${\cal D}_{nm}$, where $n,m$ are the lengths of sequences $A$ and $B$ respectively.
    - move diagonally, left or up; decision based upon the one(s) that has a maximum value of ${\cal D}_{ij}$.
    - constructs an optimally aligned sequence.
    

### Why this diversion?
* often useful to express a concept in the context of what you are already familiar with and see the connections between things that are used slightly differently or have different names.
* In this case if we changed the scoring function to find the minimum, the sequence alignmnent is finding the optimal 'path' to align two vectors within some energy landscape. This is a very generalizable notion.
* Hopefully you see that this could be applied to spell checking, and named entity identification (two aspects of NLP that are problematic internally)
    - the key here is to develop an edit distance (point mutation): either the conversion of a letter, the insertion of a letter or the deletion of a letter. This is typically done with the [Levenshtein distance](http://en.wikipedia.org/wiki/Levenshtein_distance)
    - the curious should look into these explanations: [Writing a spelling corrector](http://norvig.com/spell-correct.html), [Natural Language Corpus Data](http://norvig.com/ngrams/), [comparison of string distance algorithms](http://www.joyofdata.de/blog/comparison-of-string-distance-algorithms/), and  [Jaro-Winkler measure](http://datafireball.com/2014/10/29/jaro-winkler-string-distance/).

````python
import sys
import collections
import itertools
from scipy.stats import mode

try:
    from IPython.display import clear_output
    have_ipython = True
except ImportError:
    have_ipython = False

class KnnDtw(object):
    """K-nearest neighbor classifier using dynamic time warping
    as the distance measure between pairs of time series arrays
    
    Arguments
    ---------
    n_neighbors : int, optional (default = 5)
        Number of neighbors to use by default for KNN
        
    max_warping_window : int, optional (default = infinity)
        Maximum warping window allowed by the DTW dynamic
        programming function
            
    subsample_step : int, optional (default = 1)
        Step size for the timeseries array. By setting subsample_step = 2,
        the timeseries length will be reduced by 50% because every second
        item is skipped. Implemented by x[:, ::subsample_step]
    """
    
    def __init__(self, n_neighbors=5, max_warping_window=10000, subsample_step=1):
        self.n_neighbors = n_neighbors
        self.max_warping_window = max_warping_window
        self.subsample_step = subsample_step
    
    def fit(self, x, l):
        """Fit the model using x as training data and l as class labels
        
        Arguments
        ---------
        x : array of shape [n_samples, n_timepoints]
            Training data set for input into KNN classifer
            
        l : array of shape [n_samples]
            Training labels for input into KNN classifier
        """
        
        self.x = x
        self.l = l
        
    def _dtw_distance(self, ts_a, ts_b, d = lambda x,y: abs(x-y)):
        """Returns the DTW similarity distance between two 2-D
        timeseries numpy arrays.

        Arguments
        ---------
        ts_a, ts_b : array of shape [n_samples, n_timepoints]
            Two arrays containing n_samples of timeseries data
            whose DTW distance between each sample of A and B
            will be compared
        
        d : DistanceMetric object (default = abs(x-y))
            the distance measure used for A_i - B_j in the
            DTW dynamic programming function
        
        Returns
        -------
        DTW distance between A and B
        """

        # Create cost matrix via broadcasting with large int
        ts_a, ts_b = np.array(ts_a), np.array(ts_b)
        M, N = len(ts_a), len(ts_b)
        cost = sys.maxint * np.ones((M, N))

        # Initialize the first row and column
        cost[0, 0] = d(ts_a[0], ts_b[0])
        for i in xrange(1, M):
            cost[i, 0] = cost[i-1, 0] + d(ts_a[i], ts_b[0])

        for j in xrange(1, N):
            cost[0, j] = cost[0, j-1] + d(ts_a[0], ts_b[j])

        # Populate rest of cost matrix within window
        for i in xrange(1, M):
            for j in xrange(max(1, i - self.max_warping_window),
                            min(N, i + self.max_warping_window)):
                choices = cost[i - 1, j - 1], cost[i, j-1], cost[i-1, j]
                cost[i, j] = min(choices) + d(ts_a[i], ts_b[j])

        # Return DTW distance given window 
        return cost[-1, -1]
    
    def _dist_matrix(self, x, y):
        """Computes the M x N distance matrix between the training
        dataset and testing dataset (y) using the DTW distance measure
        
        Arguments
        ---------
        x : array of shape [n_samples, n_timepoints]
        
        y : array of shape [n_samples, n_timepoints]
        
        Returns
        -------
        Distance matrix between each item of x and y with
            shape [training_n_samples, testing_n_samples]
        """
        
        # Compute the distance matrix        
        dm_count = 0
        
        # Compute condensed distance matrix (upper triangle) of pairwise dtw distances
        # when x and y are the same array
        if(np.array_equal(x, y)):
            x_s = shape(x)
            dm = np.zeros((x_s[0] * (x_s[0] - 1)) // 2, dtype=np.double)
            
            p = ProgressBar(shape(dm)[0])
            
            for i in xrange(0, x_s[0] - 1):
                for j in xrange(i + 1, x_s[0]):
                    dm[dm_count] = self._dtw_distance(x[i, ::self.subsample_step],
                                                      y[j, ::self.subsample_step])
                    
                    dm_count += 1
                    p.animate(dm_count)
            
            # Convert to squareform
            dm = squareform(dm)
            return dm
        
        # Compute full distance matrix of dtw distnces between x and y
        else:
            x_s = np.shape(x)
            y_s = np.shape(y)
            dm = np.zeros((x_s[0], y_s[0])) 
            dm_size = x_s[0]*y_s[0]
            
            p = ProgressBar(dm_size)
        
            for i in xrange(0, x_s[0]):
                for j in xrange(0, y_s[0]):
                    dm[i, j] = self._dtw_distance(x[i, ::self.subsample_step],
                                                  y[j, ::self.subsample_step])
                    # Update progress bar
                    dm_count += 1
                    p.animate(dm_count)
        
            return dm
        
    def predict(self, x):
        """Predict the class labels or probability estimates for 
        the provided data

        Arguments
        ---------
          x : array of shape [n_samples, n_timepoints]
              Array containing the testing data set to be classified
          
        Returns
        -------
          2 arrays representing:
              (1) the predicted class labels 
              (2) the knn label count probability
        """
        
        dm = self._dist_matrix(x, self.x)

        # Identify the k nearest neighbors
        knn_idx = dm.argsort()[:, :self.n_neighbors]

        # Identify k nearest labels
        knn_labels = self.l[knn_idx]
        
        # Model Label
        mode_data = mode(knn_labels, axis=1)
        mode_label = mode_data[0]
        mode_proba = mode_data[1]/self.n_neighbors

        return mode_label.ravel(), mode_proba.ravel()

class ProgressBar:
    """This progress bar was taken from PYMC
    """
    def __init__(self, iterations):
        self.iterations = iterations
        self.prog_bar = '[]'
        self.fill_char = '*'
        self.width = 40
        self.__update_amount(0)
        if have_ipython:
            self.animate = self.animate_ipython
        else:
            self.animate = self.animate_noipython

    def animate_ipython(self, iter):
        print '\r', self,
        sys.stdout.flush()
        self.update_iteration(iter + 1)

    def update_iteration(self, elapsed_iter):
        self.__update_amount((elapsed_iter / float(self.iterations)) * 100.0)
        self.prog_bar += '  %d of %s complete' % (elapsed_iter, self.iterations)

    def __update_amount(self, new_amount):
        percent_done = int(round((new_amount / 100.0) * 100.0))
        all_full = self.width - 2
        num_hashes = int(round((percent_done / 100.0) * all_full))
        self.prog_bar = '[' + self.fill_char * num_hashes + ' ' * (all_full - num_hashes) + ']'
        pct_place = (len(self.prog_bar) // 2) - len(str(percent_done))
        pct_string = '%d%%' % percent_done
        self.prog_bar = self.prog_bar[0:pct_place] + \
            (pct_string + self.prog_bar[pct_place + len(pct_string):])

    def __str__(self):
        return str(self.prog_bar)
````
