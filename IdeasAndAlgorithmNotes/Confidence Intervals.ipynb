```python
from IPython.display import Math
import numpy as np
import scipy
import scipy.stats as stats

pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

## Data

```python
np.random.seed(7654567) 

def genData(l,s,n,goal):
    rvs = stats.norm.rvs(loc=l, scale=s, size=n)
    
    while any(rvs<0) or any(rvs>10) :
        rvs[rvs<0]=stats.norm.rvs(loc=l, scale=s, size=count_nonzero(rvs<0))
        rvs[rvs>10]=stats.norm.rvs(loc=l, scale=s, size=count_nonzero(rvs>10))
    
   
    xbar=np.mean(rvs)
    s=np.std(rvs)
    n=len(rvs)
    mu=goal
    return (rvs,xbar,s,n,mu)
rvs,xbar,s,n,mu = genData(3,1,500,5)

def plotHist():
    hN,hBins,hPatches = hist(rvs,color='lightblue')
    hMean = axvline(xbar,color='b', linewidth=2, linestyle='--')
    hGoal = axvline(mu,color='r', linewidth=4, linestyle='-')
    legend((hMean,hGoal), ('Mean='+str(round(xbar,2)),'Goal='+str(mu)))
    xlim(0,10)

print "Sample mean:",xbar
print "Sample std:",s
print "Sample size:",n
print "Goal:",mu
print

plotHist()
```

## T-statistic method

```python
def tStat():
    rv = stats.t(n-1)
    
    t = (xbar-mu)/(s/sqrt(n))
    
    print
    print "----- T-statistic metric -----"
    
    print "T stat:",t
    
    
    tCDF = rv.sf(t)
    
    print "P[t<T]=",tCDF
    print
    
    tval = rv.isf(0.05/2)
    print "t-val: ", tval
    
    print "95% CI:", (xbar - tval * s/sqrt(n),xbar + tval * s/sqrt(n))
    
    osTval=rv.ppf(0.05)
    print
    print "one-sided t-val:",osTval
    print "95% lower confidence bound (P[X<metric]=0.95): X = ", (xbar - osTval * s/sqrt(n))
    
    print
    
    print "Prob that metric is met or exceded:",1-tCDF
    
    
    mv = rv.ppf(0.001)
    mv = np.min([-np.abs(t)*1.1,mv])
    
    x=np.linspace(mv,-mv,1000)
    x=append(x,t)
    x=sort(x)
    
    y=rv.pdf(x)
    
    figure()
    subplot(122)
    
    hPlot=plot(x,y)
    hFill=fill_between(x,0,y, x<=t, facecolor='lightgreen', interpolate=True)
    hLine, =plot([t, t],[0, rv.pdf(t)],'r',label="T-statistic="+ str(np.round(t,3)))
    legend()
    p = Rectangle((0, 0), 1, 1, fc="lightgreen")
    legend([hLine,p],["T-statistic="+ str(np.round(t,3)),"P[success]="+str(np.round(1-tCDF,3))])
    
    h=title('Sudent\'s t distribution')
    ylim(0,1)
    
    subplot(121)
   
    plotHist()
    return 1-tCDF


tStat()
```

## Approach using CDF of fitted probability distribution

```python
def normFit():
    rv = stats.norm(loc=xbar,scale=s)
    t = rv.pdf(mu)
    
    print
    print "----- Normal PDF fit metric -----"
    print "PDF of distribution at goal:",t
    
    tCDF = rv.cdf(mu)
    
    print "P[T<mu]=",tCDF
    print "P[mu<T]=",1-tCDF
    
    print "Success=",(1-tCDF)/tCDF
    print
    
    
    mv = rv.ppf(0.001)
    #mv = np.min([np.abs(t)*1.1,mv])
    
    x=np.linspace(rv.ppf(0.001),rv.ppf(0.999),1000)
    x=append(x,t)
    x=sort(x)
    
    y=rv.pdf(x)
    
    figure()
    subplot(122)
    ax=gca()
    hPlot=plot(x,y)
    hFill=fill_between(x,0,y, x>=mu, facecolor='lightgreen', interpolate=True)
    hFill=fill_between(x,0,y, x<mu, facecolor='pink', interpolate=True)
    hLine, = plot([mu, mu],[0, t],'r')
    hXbar, = plot([xbar, xbar],[0, rv.pdf(xbar)],'b--')
    
    p = Rectangle((0, 0), 1, 1, fc="lightgreen")
    q = Rectangle((0, 0), 1, 1, fc="pink")
    legend([hLine,p,q],["Goal="+ str(np.round(mu,3)),"P[answer>goal]="+str(np.round(1-tCDF,3)),"P[answer<goal]="+str(np.round(tCDF,3))])
    
    
    ax.text(.1,.8, 'Score='+str(round((1-tCDF)/tCDF,3)), fontsize=14, bbox=dict(facecolor='gray', alpha=.25),
             transform=ax.transAxes)
    h=title('Normal distribution')
    ylim(0,1)
    
    subplot(121)
    
    plotHist()
    return (1-tCDF)/tCDF
    
normFit()
```
