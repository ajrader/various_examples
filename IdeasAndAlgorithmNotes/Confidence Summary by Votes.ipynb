<H1> Voting approach to defining Confidence Intervals</H1>
<p> Much of the motivation here stems from not trusting that the survey results 
would necessarily be normally distributed. This approach will assign the three color values: 
    red (bad), yellow (marginal), and green (good) to each survey result and then aggregate them.</P>

<p> tricolor is a list of the 3 key colors: red, yellow and green (html codes).</p>

```python
import scipy
import scipy.stats as stats

pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
np.random.seed(7654567) 

def genData(l,s,n,goal):
    rvs = stats.norm.rvs(loc=l, scale=s, size=n)
    
    while any(rvs<0) or any(rvs>10) :
        rvs[rvs<0]=stats.norm.rvs(loc=l, scale=s, size=count_nonzero(rvs<0))
        rvs[rvs>10]=stats.norm.rvs(loc=l, scale=s, size=count_nonzero(rvs>10))
    # convert to integers:
    rvs = rvs.astype(int)
    xbar=np.mean(rvs)
    s=np.std(rvs)
    n=len(rvs)
    mu=goal
    return (rvs,xbar,s,n,mu)
rvs,xbar,s,n,mu = genData(3,1,500,5)

def plotHist():
    hN,hBins,hPatches = hist(rvs,color='lightblue')
    hMean = axvline(xbar,color='b', linewidth=2, linestyle='--')
    hGoal = axvline(mu,color='r', linewidth=4, linestyle='-')
    legend((hMean,hGoal), ('Mean='+str(round(xbar,2)),'Goal='+str(mu)))
    xlim(0,10)

print "Sample mean:",xbar
print "Sample std:",s
print "Sample size:",n
print "Goal:",mu
print

plotHist()

def plotHist2(rvs,xbar,std,mu,n):
    hN,hBins,hPatches = hist(rvs,color='lightblue')
    hMean = axvline(xbar,color='b', linewidth=2, linestyle='--')
    hGoal = axvline(mu,color='r', linewidth=4, linestyle='-')
    legend((hMean,hGoal), ('Mean='+str(round(xbar,2)),'Goal='+str(mu)))
    xlim(0,10)
```

<h2> Look at poisson distribution instead</h2>

```python
def genDataPoisson(l,n,goal):
    rvs = stats.poisson.rvs(l, size=n)
    
    while any(rvs<0) or any(rvs>10) :
        rvs[rvs<0]=stats.norm.rvs(l, size=count_nonzero(rvs<0))
        rvs[rvs>10]=stats.norm.rvs(l, size=count_nonzero(rvs>10))
       
    xbar=np.mean(rvs)
    s=np.std(rvs)
    n=len(rvs)
    mu=goal
    return (rvs,xbar,s,n,mu)

Prvs,Pxbar,Ps,Pn,Pmu = genDataPoisson(3,500,5)
```

```python
def distribVoting(rvs,goal,n,acceptWindow=1,margWindow=2):
    import numpy as np
    minResponse = 0
    maxResponse = 10
    x = np.arange(minResponse,maxResponse+1)
    ctmp = ['yellow']*len(x)
    ccount=np.zeros(len(x))
    for i in x:
        ccount[i] = len(rvs[rvs == i])
        if i < (goal - acceptWindow -margWindow):
            ctmp[i] = 'red'
        elif i >= goal - acceptWindow:
            ctmp[i] = 'green'
        #print i,ccount, ctmp[i]
        
    good = len(rvs[rvs >= (goal - acceptWindow)])
    bad = len(rvs[rvs < (goal - acceptWindow - margWindow)])
    okay = n - good - bad
    
    
    Score = (1.*bad+2.*okay + 3.*good)/float(n)
    ct=assignConsensusColor(Score)
    #subplot(111, axisbg = ct)
    maxHeight = max(ccount)
    xlim(0,10)
    bar(x,ccount,color=ctmp)
    text(1,maxHeight*0.8,'Score={0}'.format(Score),backgroundcolor=ct)
    
    
    return (bad, okay, good,Score)

def assignConsensusColor(Score, colorStride=4):
    y2g = ['#FFFF00','#F2F900','#E6F200','#D9EC00','#CCE600','#BFDF00','#B2D900','#A6D300','#99CC00','#8CC600',
       '#80C000','#73B900','#66B300','#59AC00','#4DA600','#40A000','#339900','#269300','#198D00','#0D8600','#008000']
    r2y = ['#FF0000','#FF0D00','#FF1A00','FF2600','#FF3300','#FF4000','#FF4C00','#FF5900','#FF6600','#FF7300','#FF8000',
 	   '#FF8C00','#FF9900','#FFA600','#FFB200','#FFBF00','#FFCC00','#FFD900','#FFE600','#FFF200',  '#FFFF00']
    #xx = np.arange(0,len(sNorm))
    
    stest = Score-1
    ncbins = len(r2y[::colorStride])
    if stest < 1:
        crange = r2y[::colorStride]
        cval = int(ncbins*stest)
    
    else:
        crange = y2g[::colorStride]
        cval = int(ncbins*(stest -1))
#        print "**\t",cval, crange[cval]#int(ncbins*(stest-1))       
  
    return (crange[cval])
    
```
