README:

    This .ipynb file can be used to help grade applicant assignments. Running this script will open all of the applicants submissions, plot them, and provide a few metrics of accuracy. The script will also rank the applicants based on these metrics (from best to worst). 
    
    This notebook is designed to be started from the same folder that the submissions files are in.  The easiest way to use this notebook is:
    
    1. Put all submission folders into the same folder
    2. Start this notebook from that folder
    3. Change the "holdout_data_path" variable in the 3rd cell to the path of your holdout.csv file
    4. Run the notebook
    
    If step (2) cannot be done, then you will need to set the working directory of the notebook to be the folder set in step (1)
    
    
    
    


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.metrics import mean_squared_error
from scipy.stats import ks_2samp
import re
%matplotlib inline
```

```python
####Change this to reflect whatever directory your holdout.csv file is in####

holdout_data_path = "Holdout.csv"
```

```python
def diagnostic_printer(y,yhat,applicant_name):
    RMSE = np.sqrt(mean_squared_error(y,yhat))
    ks = ks_2samp(y,yhat)
    
    print "Results for "+ applicant_name
    print "RMSE - ", RMSE
    print "KS-test Statistic - ", ks[0]
    
    return RMSE, ks[0], applicant_name
```

```python
def plot_data(y,y_predict, y_label='y', y_pred_label='y_hat',title = 'Needs Name'):
    h = plt.figure(figsize=(20,10))
    sns.set_style("whitegrid")
    sns.distplot(y_predict, hist=False , label=y_pred_label)
    sns.distplot(y,hist=False, label=y_label)
    plt.legend()
    plt.show()
```

```python
from glob import glob
```

```python
####Initialization#####

##Get names of folders
folder_names = glob('./*/Predictions.csv')

##Make sure we only have submission files
submission_names = folder_names

#Load in the holdout
holdout = pd.read_csv(holdout_data_path)
holdout = holdout[holdout.Interest_Rate.notnull()]
holdout['y'] = holdout.Interest_Rate.apply(lambda x: float(x.replace(r'%','')))

##The file with the predictions could be called "Predictions.csv"
# 2 columns, 1 that is first model prediction, 2 that is second model prediction
# should just be floating number, no % or anything.

scores =[]

for folder in submission_names:
    
    print "================================="
    print "######## New Applicant ##########"
    print "================================="
    
    #Just get applicants name
    app_name = folder.replace('.\\','').replace('./','').split('\\')[0].split('/')[0]
    
    #Get the path of the Predictions.csv file
    prediction_name = folder
    print prediction_name
    
    #Read in the file
    predictions = pd.read_csv(prediction_name)
    
    predictions.columns = ['First', 'Second']
    for column in predictions.columns:
        if predictions[column].dtype == 'object':
            predictions[column] = predictions[column].apply(lambda x: float(x.replace(r'%','')))
    predictions.fillna(predictions.mean(), inplace=True)
    
    
    #Get length of submitted predictions
    prediction_length = len(predictions)
    
    #Only take set of holdout that matches with length of submission
    #Also fix scale issues
    if np.median(predictions.First) < 1:
        predictions.First = predictions.First * 100.0
        predictions.Second = predictions.Second * 100.0
        
    y = holdout.y[:prediction_length]
    
    #First submission RMSE/KS
    diag = diagnostic_printer(y,predictions.First,app_name)
    
    #First submission plot
    plot_data(y,predictions.First)
    #Save
    scores.append(diag)

    #Second submission RMSE/KS
    diag = diagnostic_printer(y,predictions.Second,app_name)

    #Second submission plot
    plot_data(y,predictions.Second)
    scores.append(diag)

print '======================================='    
print "Rankings"
print '======================================='
#Get dataframe of scores
scores = pd.DataFrame(scores,columns=['RMSE','KS','Applicant'])
#Sort based on RMSE
print 'Sorted RMSE'
print scores.sort(columns='RMSE')
print '======================================='
#Sort based on KS
print 'Sorted KS-Statistic'
print scores.sort(columns='KS')
```
