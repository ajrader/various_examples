```python
# 2. process seqdumper file
# Func to load recommendations file

def loadRecommends(fileName):
    infile =  open(fileName,'r')
    k=0
    recommendations = {}
    p= re.compile('^(\d+)\t\[(\S+)\]$')
    for line in infile:
        m = p.match(line)
    #print len(m.groups()),m.group(1), m.group(2)
        rvals = m.group(2).split(',')
        myReco = {}
    
        for r in rvals:
            rr = r.split(':')
            myReco[rr[0]] = rr[1]
    
        recommendations[m.group(1)] = myReco
    
    infile.close()
    return recommendations
    
#function to parse a sequencedumper KV file:
def convertSeqKVfile(filename):
    ifile = open(filename,'r')
    myFeature = {}
    pKV=re.compile('^Key:\s(\d+):\sValue:\s\{(\S+)\}$')
    for line in ifile:
        m=pKV.match(line)
        if m is None: continue
    #    print m.group(1), m.group(2)
        vals = m.group(2).split(',')
        myDim = {}
        for v in vals:
            v2 = v.split(':')
            myDim[int(v2[0]) ] = float(v2[1])

        myFeature[m.group(1)] = myDim
    ifile.close()
    return myFeature
    
```
