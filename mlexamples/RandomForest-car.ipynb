#### change the category names to floats:

```python
carG = carRaw
i=0
for a in carRaw.buying.unique():
    tmp = carRaw.buying == a
    carG.buying[tmp] = float(i)
    i+=1
    
carG.buying.value_counts()
```

```python
i = 0
for a in carRaw.maint.unique():
    tmp = carRaw.maint == a
    carG.maint[tmp] = float(i)
    i+=1

i=0
for a in carRaw.doors.unique():
    tmp = carRaw.doors == a
    carG.doors[tmp] = float(i)
    i+=1
    
i=0
for a in carRaw.persons.unique():
    tmp = carRaw.persons == a
    carG.persons[tmp] = float(i)
    i+=1
    
i=0
for a in carRaw.lug_boot.unique():
    tmp = carRaw.lug_boot == a
    carG.lug_boot[tmp] = float(i)
    i+=1
    
i=0
for a in carRaw.safety.unique():
    tmp = carRaw.safety == a
    carG.safety[tmp] = float(i)
    i+=1

carG.maint.value_counts(), carG.doors.value_counts(),carG.persons.value_counts(),carG.lug_boot.value_counts(),carG.safety.value_counts()
```

```python
N = len(carG)
N
```

```python
i=0
for a in carRaw['class'].unique():
    tmp = carRaw['class'] == a
    carG['class'][tmp] = float(i)
    i+=1
```

```python
carG['class'].value_counts()
```

### Ready to test Random Forest: Divide into test and train by 70/30 split

```python
ntest = int(.3*N)
ntrain = N-ntest
ntrain,ntest
```

##### use scikit-learn for defining test and train sets  convert carG.class to target and rest to data: (N,) and (N,6)

```python
from sklearn import cross_validation
cdata = carG.as_matrix(['buying','maint','doors','persons','lug_boot','safety'])
target = carG.as_matrix(['class'])
```

```python
X_train, X_test, y_train,y_test = cross_validation.train_test_split(cdata,target,test_size=0.3)
```
