<H1> Case Study: LOESS with matplotlib</H1>
    <H3> Example from Data Analysis with Open Source Tools by P.K. Janert</H3>
    <p> Recall that LOESS means <em> locally weighted linear regression</em>. This means there is a 
    local <em> weight factor</em> applied to the data points that are particularlly close to a given $x$.
    The expression for squared error (the function we want to minimize) is 
    $$ \chi^2(x) = \sum_i w(x-x_i;h) (a+bx_i-y_i)^2 $$
    which depends explicitly upon x, the location where we want to evalutate the smoothed curve.</p>
    <p> Minimization of this function with respect to the two parameters, $a$ and $b$, yields:
    $$ b = \frac{\sum w_i\sum w_ix_iy_i - (\sum w_i x_i)(\sum w_iy_i)}{\sum w_i(\sum w_i x_i^2) - (\sum w_i x_i)^2}$$
    and 
    $$ a = \frac{\sum w_iy_i - b\sum w_i x_i}{\sum w_i}$$
    </p>
    </p> We will encapsulate this into the function <code>loess()</code></p>

```python
%pylab inline
import pandas as pd
```

```python
# x: location; h: bandwith; xp, yp: data points (vectors)
def loess(x,h,xp,yp):
    w = exp(-0.5*( ((x-xp)/h)**2) /sqrt(2*pi*h**2))
    b = sum(w*xp)*sum(w*yp) - sum(w)*sum(w*xp*yp)
    b /= sum(w*xp)**2 - sum(w)*sum(w*xp**2)
    a = ( sum(w*yp) - b* sum(w*xp) )/sum(w)
    
    return a+ b*x
    
            
```

```python
#load the auto-mpg data file
autoFileName = '../data/auto-mpg.data'
cwidths = [7,4,10,10,11,7,4,4,30]
cnames = ['mpg','cyl','displ','hp','wt','accel','model_year','origin','carName']
d = pd.read_fwf('../data/auto-mpg.data',header=None,widths=cwidths,na_values=['?'],names=cnames)
print len(d)
```

```python
#sort the dataframe by mpg
d2 = d.sort(['mpg'],ascending=[True])
x = d2.mpg
y = d2.hp
#y = d.mpg # assign y to mpg
#x=d.hp #assign x to horespower (hp)
```

```python
# create lists for fits
s1, s2 = [],[]
for k in x:
    s1.append(loess(k,0.5, x,y))
    s2.append(loess(k,1.0, x, y))
```

```python
figure(figsize=(10,8))
xlabel("Horsepower",fontsize=14)
ylabel("MPG",fontsize=14)
plot( x,y, 'o',color='steelblue',alpha=0.5,markersize=7, linewidth=3)
plot(x,array(s1),'k-',x,array(s2),'r--',lw=2)
grid(True)
```

```python
from scipy.stats.kde import gaussian_kde
from scipy.stats import norm
from numpy import linspace,hstack
from pylab import plot,show,hist

# creating data with two peaks
sampD1 = norm.rvs(loc=-1.0,scale=1,size=300)
sampD2 = norm.rvs(loc=2.0,scale=0.5,size=300)
samp = hstack([sampD1,sampD2])

# obtaining the pdf (my_pdf is a function!)
my_pdf = gaussian_kde(samp)

# plotting the result
x = linspace(-5,5,100)
plot(x,my_pdf(x),'r') # distribution function
hist(samp,normed=1,alpha=.3) # histogram
show()
```
