```python
from itertools import chain
import re
import pandas as pd

def stepThroughDetailedRecord(detailRecord):
    sellerComments = ''
    stdEquip = [] #empty list of standard equipment
    # input is a beautifulSoup record of the CARS.COM Details Page
    # example page: http://www.cars.com/vehicledetail/detail/614932873/overview/
    # Step 1 identify the Vehicle Details; returns a dictionary
    drow = detailRecord.find('div',{'class':'col49 vehicle-data'}) # identify the listing of vehicle-data
    vehicleDetailsList = drow.find_all('ul', {'class': 'vehicle-details list'})
    #print len(vehicleDetailsList)
    vehDetails = convertVehicleDetailItems(vehicleDetailsList)
    
    # Step 2 get seller comments
    # copy all the seller's comments
    sellerComments = detailRecord.find('span',{'class':'seller-notes'}).text
    # clean up the comments --> remove extra non-alpha numeric info
    
    myS = re.sub('[^A-Za-z0-9!\?\.#]+', ' ', sellerComments)
    if myS.endswith(' Show More '):
        myS = myS[:-11]
    sellerComments = myS
    # Step 3 get standard Equipment information
    stdEquipRec = detailRecord.find('ul',{'class':'st-equipment list'})
    for c in stdEquipRec.children:
        try:
            stdEquip.append(c.text.split('\n'))
        except:
            pass
        
    # clean-up this list (flatten and remove unnecessary info)
    #print len(stdEquip)
    stdEquip = list(chain.from_iterable(stdEquip))
    stdEquip.remove(' Show More')
    stdEquip = filter(None, stdEquip)
    #print len(stdEquip)
    # Step 4 get  Dealer Info
    
    drow = detailRecord.find('div',{'class':'col49 about-dealer'}) # identify the listing of vehicle-data
    dealerDiv = drow.find('div', {"class":"col27"})
    dealerDetails = returnDealerInfo(dealerDiv)
    # step 5 push dictonaries into a pandas data frame
    # generates a tuple of a dictonary, string, list and dictonary: (vehDetails, sellerComments, stdEquip, dealerDetails)
    t1 = pd.DataFrame(vehDetails.values())
    t2 = pd.DataFrame(dealerDetails.values())
    cnames = vehDetails.keys() + dealerDetails.keys() #+ ['StdEquipList','comments']
    vehicleInfo = pd.concat([t1.T,t2.T],axis=1)
    vehicleInfo.columns = cnames
    # returns a dataframe, list and string
    return(vehicleInfo,stdEquip,sellerComments)
    #return((vehDetails,sellerComments,stdEquip,dealerDetails))

    
def convertVehicleDetailItems(vehicleDetails):
    #vehicleDetailsList = drows[7].find_all('ul', {'class': 'vehicle-details list'})
    mySplit = vehicleDetails[0].text.split('\n')[:-2]# drop the last 2 (empty items alternatively learn to parse these
      
    nkeys = (len(mySplit)-1)/2
    myDetailItems = {}
    #omit first entry
    keys = [a for a in mySplit[1::2]]
    values = [b for b in mySplit[2::2]]
    for k in xrange(nkeys): # fix the length to deal if they are longer
        myDetailItems[keys[k]]=values[k]

    return myDetailItems

def returnDealerInfo(dealerDiv):
    myDealerKeys = {'name','href','Nratings','Rating','StreetAddress','City','State','Zip','website','phone'}
    myDealerInfo ={}
    mname = dealerDiv.find('a')
    myDealerInfo['name'] = mname.text
    myDealerInfo['href'] = mname.get('href')
    mrating = dealerDiv.find('div',{'class':'detail'})
    myDealerInfo['Nratings'] = mrating.find('a').text
    ratingRaw = mrating.text.split('\r')[0]
    myDealerInfo['Rating'] = re.sub('[^0-9\.]+','',ratingRaw)
    maddress = dealerDiv.find('p',{'class':'address'})
#print address.prettify()
    address = maddress.text.split('\r\n')
    myDealerInfo['StreetAddress'] = address[0] # first element of this split
    CityStateZip = address[1].split(',')
    myDealerInfo['City'] = CityStateZip[0].strip(' ')
    myTrash,myDealerInfo['State'],myDealerInfo['Zip'] = CityStateZip[1].split(' ')
    #myDealerInfo['address']=maddress.text.split('\r\n')
    
    phoneRaw = dealerDiv.find('p',{'class':"phone"}).text
    myDealerInfo['phone']  = re.sub('[^A-Za-z0-9\(\)\-]+','',phoneRaw)
    # get website:
    jj = dealerDiv.find('p',{'class':"links"})
    for link in jj.find_all('a'):
        if ('Visit dealer site' in link.text):
            myDealerInfo['website'] = link.get('href')
    
    return myDealerInfo
# returns a dictionary of the vehicle details/specs    
```
