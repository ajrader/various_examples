## Goal: come up with a $\it{telematic fingerprint}$ to predict if a trip was given by a driver.

For automobile insurers, telematics represents a growing and valuable way to quantify driver risk. Instead of pricing decisions on vehicle and driver characteristics, telematics gives the opportunity to measure the quantity and quality of a driver's behavior. This can lead to savings for safe or infrequent drivers, and transition the burden to policies that represent increased liability.

AXA has provided a dataset of over 50,000 anonymized driver trips. The intent of this competition is to develop an algorithmic signature of driving type. Does a driver drive long trips? Short trips? Highway trips? Back roads? Do they accelerate hard from stops? Do they take turns at high speed? The answers to these questions combine to form an aggregate profile that potentially makes each driver unique.

For this competition, Kaggle participants must come up with a "telematic fingerprint" capable of distinguishing when a trip was driven by a given driver. The features of this driver fingerprint could help assess risk and form a crucial piece of a larger telematics puzzle.

## Featurs of the dataset
* directories for different drivers
* 200 different files for 200 trips within each of these directories
* format of the files is x,y representing the car positions (in meters) every second:
    ` x,y `
    ` 0.0,0.0
    `18.6,-11.1`
    `36.1,-21.9`
    `...`
* all trips centered to start at the origin(0,0), randomly rotated, and short lengths of trip data were removed from the start/end of the trip.
* a small and random number of false trips are planted in each driver's folder.
    * the number of these false trips varies
    * these false trips are not labeled (i.e. no True Positives/ False Positives are known)
    * majority of the trips in each folder do belong to the same driver

```python
def calc_tripDF(trip_list):
    d1trips = pd.DataFrame(columns=['Nsegments','totalDistance','medianSpeed','meanSpeed','meanAccel','medianAccel'])
    i=0
    for a in trip_list:
        tripDF = pd.read_csv(a)
        nsegs = len(tripDF)
    
        distances = tripDF.diff().apply(lambda x: x**2).sum(axis=1).apply(lambda x: np.sqrt(x))
        totalDist = distances.sum()
        medianSpeed = distances.median()
        meanSpeed = distances.mean()
        accelerations = distances.diff()
        meanAccel = accelerations.mean()
        medianAccel = accelerations.median()
        mydata = [nsegs,totalDist,medianSpeed,meanSpeed,meanAccel,medianAccel]
        #print i,nsegs,totalDist,medianSpeed,meanSpeed,meanAccel,medianAccel
        d1trips.loc[i,:]= mydata
        i+=1

    #d1trips.head()
    for col in d1trips.columns:
        col_zscore = col + '_zscore'
        d1trips[col_zscore] = (d1trips[col] - d1trips[col].mean())/d1trips[col].std(ddof=0)

    return d1trips
```

```python
def calc_prob(x):
    prob = 2*stat.norm.cdf(-abs(x))
    return prob
```

```python
def calc_bin_prob(x,zcutoff):
    prob = 1
    if abs(x) >= zcutoff:
        prob = 0
    return prob
```
